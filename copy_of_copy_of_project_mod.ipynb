The harmonic oscillator wave functions are given by,

 $\Psi _{n} = A_{n}H_{n}(x)e^{-x^{^{2}/2}}$

where, $A_{n}$ is the normalization constant (ignore it here), $H_{n}(x)$ is the Hermite polynomial of order $n$, and $x$ is the dimensionless coordinate. The Hermite polynomial is given by,
$H_{n}=n!\sum_{l=0}^{n/2}\frac{(-1)^{n/2-l}}{(2l)!(n/2-l)!}(2x)^{2l}\: \: \:\;  for\:  even \: n$
$H_{n}=n!\sum_{l=0}^{(n-1)/2}\frac{(-1)^{\frac{n-1}{2}-l}}{(2l+1)!(\frac{n-1}{2}-l)!}(2x)^{2l+1}\: \: \:\;  for\:  odd \: n$


````python
import matplotlib.pyplot as plt
def fac(n):
  fact=1
  for x in range(1,n+1):
    fact=fact*x
  return fact

def Hn_odd(n,x):
  s=0
  lmax=int((n-1)/2)
  for l in range (0,lmax+1):
   s1 = ((-1)**(lmax-l)*(2*x)**(2*l+1))/(fac(2*l+1)*fac(lmax-l))
   s += s1
  return s

def Hn_even(n,x):
  s=0
  lmax=int(n/2)
  for l in range (0,lmax+1):
   s2 = ((-1)**(lmax-l)*(2*x)**(2*l))/(fac(2*l)*fac(lmax-l))
   s += s2
  return s

import numpy as np
def gau(x):
  b=np.exp((-x**2)/2)
  return b
  
def psi(n,x):
  if n % 2==0:
    p= fac(n)*Hn_even(n,x)*gau(x)
    return p
  else:
    q= fac(n)*Hn_odd(n,x)*gau(x)
    return q

c=[]
d=[]
e=[]
f=[]
n=int(input("Place the 'n' value: "))
for i in range (-5,6):
  z= psi(n,i)
  c.append(i)
  d.append(z)
  e.append(i)
  f.append(z**2)

plt.subplot(1,2,1)
plt.plot(c,d,"go-")
plt.xlabel("$x$")
plt.ylabel("$\psi_{n}(x)$")
plt.subplot(1,2,2)
plt.plot(e,f,"y*-")
plt.xlabel("$x$")
plt.ylabel("$|\psi_{n}(x)|^{2}$")
````

````python
import math

def hermite(x, n):
    xi = np.sqrt(m*w/hbar)*x
    herm_coeffs = np.zeros(n+1)
    herm_coeffs[n] = 1
    return Herm.hermval(xi, herm_coeffs)
  
def stationary_state(x,n):
    xi = np.sqrt(m*w/hbar)*x
    prefactor = 1./math.sqrt(2.**n * math.factorial(n)) * (m*w/(np.pi*hbar))**(0.25)
    psi = prefactor * np.exp(- xi**2 / 2) * hermite(x,n)
    return psi
n=int(input("Place the 'n' value: "))
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.plot(x, stationary_state(x,n))
plt.xlabel("x")
plt.ylabel("$\psi_n(x)$")
plt.title("Test Plot of $\psi_n(x)$")
plt.subplot(1,2,2)
plt.plot(x, (stationary_state(x,n))**2)
plt.xlabel("x")
plt.ylabel("$|\psi_n(x)|^{2}$")
plt.title("Test Plot of $|\psi_n(x)|^{2}$")
plt.show()
````

Let's find the harmonic oscillator wavefunctions and energy eigenvalues from Time independent Schrodinger Equation by solving it $Numerically$.

````python
# make a simple animation
!pip install celluloid 

from celluloid import Camera # getting the camera
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
````

````python

fig, ax = plt.subplots(figsize=(10, 6))
camera = Camera(fig)
# define parameters 
hbar, m, l, N = 1,1,5,100
# define grid and potential energy
x = np.linspace(-l,l,N)
delx = x[1]-x[0]
alpha = - hbar**2/(2*m*delx**2)
V = 0.5*x**2  # for SHO

# define T_matrix and V_matrix
V_matrix = np.zeros((N,N))
T_matrix = np.zeros((N,N))

# V_matrix construction
for i in range(N):
  V_matrix[i,i] = V[i]

# T_matrix construction
for i in range(N):
  T_matrix[i,i] = -2.0*alpha
  if i < N-1:
    T_matrix[i,i+1] = 1.0*alpha
  if i > 0:
    T_matrix[i,i-1] = 1.0*alpha

# H_matrix
H_matrix = V_matrix + T_matrix

# diagonalize H_matrix
eigval,eigvec = np.linalg.eig(H_matrix)
idx = eigval.argsort()
eigval = eigval[idx]
eigvec = eigvec[:,idx]
for i in range(6):
  print(i+1,"Relative energy for n=",i,":",eigval[i]/eigval[0])
plt.subplot(1,2,1)
for i in range(6):
  plt.plot(x,(eigvec[:,i]),label=i)
  plt.xlabel("x")
  plt.ylabel("$\psi_n(x)$")
  plt.title("Test Plot of $\psi_n(x)$")
  plt.axhline()
  plt.legend()
  camera.snap()
plt.subplot(1,2,2)
for i in range(6):
  plt.plot(x,(eigvec[:,i])**2,label=i)
  plt.xlabel(r"x")
  plt.ylabel(r"$|\psi_n(x)|^{2}$")
  plt.title(r"Test Plot of $|\psi_n(x)|^{2}$")
  plt.axhline()
  plt.legend()
  camera.snap()

animation = camera.animate()
html=display.HTML(animation.to_html5_video())
display.display(html)
plt.close()
````

Now returns the classical probability density for a particle with an energy equal to that of the $n^{th}$ quantum harmonic oscillator energy level.
For a classical harmonic oscillator with energy  $E$ , the classical probability of finding the particle at  $x$  is given by,

$x_{max}=\sqrt{\frac{2E}{m\omega ^{2}}}$

where,  $x_{max}$  is the classical turning point, and  $P_{classical(x)}$  is understood to be zero for  $|x|>|xmax|$ .

````python

#Discretized space
dx = 0.05
x_lim = 12
x = np.arange(-x_lim,x_lim,dx)

def classical_P(x,n):
    E = hbar*w*(n+0.5)
    x_max = np.sqrt(2*E/(m*w**2))
    classical_prob = np.zeros(x.shape[0])
    x_inside = abs(x) < (x_max - 0.025)
    classical_prob[x_inside] = 1./np.pi/np.sqrt(x_max**2-x[x_inside]*x[x_inside])
    return classical_prob
n=int(input("Place the 'n' value: "))
plt.figure()
plt.plot(x, classical_P(x,n))
plt.xlabel("x")
plt.ylabel("$P_{classical}(x,n)$")
plt.title("Classical Probability Density of a Harmonic Oscillator, $E=E_n$")
plt.show()
````

Now we are in a place to compare this classical and quantum behaviour of harmonic oscillator.

The classical probability distribution says that there is some chance of finding the particle at any point within the classically allowed region. In contrast, the quantum probability distributions possess nodes of zero probability within this region.

For classical oscillator the probability density is maximum near the classical turning points. But for quantum oscillator this is not the case. for example, at ground state probability density is maximum at $x=0$. But from the above animation it is obvious that as $n\rightarrow $ higher quantum number the probabilty density plot looks similar. This can be interpretated from $'Bohr \:\:correspondence\:\:principle'$.

Let's check now if there is any probability of finding the particle outside the classical turning points.

````python
def simpson_rule(y,h):
  sum = y[0]+y[-1]
  for i in range (1,len(y)-1):
    if i % 2 ==0:
      ai =2
    else:
      ai =4
    sum += ai*y[i]
  sum *= h/3
  return sum
n=int(input("Place the eigenstate: "))
E = hbar*w*(n+0.5)
x_max = numpy.sqrt(2*E/(m*w**2))
#print("Magnitude of classical turning points:",x_max)
z=np.linspace(-x_max,x_max,200)
h=z[1]-z[0]
intg= simpson_rule((stationary_state(z,n))**2,h)
print("Probability of finding particle inside the harmonic potential:",intg)
tun=1-intg
print("Tunnelling probability:",tun)
````

Oh! There is non zero probability of finding the particle in the classically forbidden zeros. So, there is $'Tunnelling'!$ for each value of chosen eigenstate. For example, for the ground state there is almost $16$% chance of tunnelling.

````python
for i in range (0,40,2):
  E = hbar*w*(i+0.5)
  x_max = numpy.sqrt(2*E/(m*w**2))
  z=np.linspace(-x_max,x_max,200)
  h=z[1]-z[0]
  intg= simpson_rule((stationary_state(z,i))**2,h)
  #print("Probability of finding particle inside the harmonic potential:",intg)
  tun=1-intg
  print("Tunnelling probability:",tun,"for n=",i)
````

The values indicate that as $'n'$ value increases tunnelling probability decreases. This is also a consequence of $'Bohr \:\:correspondence\:\:principle'$.
